```python
import opendatasets as od
import os
import pandas as pd
```

```python
# Descarga de los datos desde Kaggle
# {"username":"miguelgallego","key":"6a7a2546d0b0a5683db1abd67fc0380c"}
dataset_link = "https://www.kaggle.com/datasets/sainikhileshreddy/food-recognition-2022"
od.download(dataset_link)

# os.chdir("food-recognition-2022")
# os.listdir()
```

```python
# Ruta de la carpeta principal del conjunto de datos
dataset_dir = "/content/food-recognition-2022"

# Lista de carpetas dentro de la carpeta principal
subdirectories = os.listdir(dataset_dir)
print("Contenido de la carpeta principal:")
print(subdirectories)
```

```python
import cv2
import matplotlib.pyplot as plt

# Ruta de la carpeta de entrenamiento
train_dir = os.path.join(dataset_dir, "raw_data/public_training_set_release_2.0/images")

# Tomar algunas imágenes de ejemplo
sample_images = os.listdir(train_dir)[:1]  # Toma las primeras 5 imágenes como ejemplo

# Mostrar las imágenes de ejemplo
for image_filename in sample_images:
    image_path = os.path.join(train_dir, image_filename)
    image = cv2.imread(image_path)
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(f"Imagen: {image_filename}")
    plt.show()
```

```python
import json

# Ruta al archivo JSON de anotaciones
annotations_file = "/content/food-recognition-2022/raw_data/public_training_set_release_2.0/annotations.json"

# Cargar el archivo JSON
with open(annotations_file, "r") as json_file:
    annotations_data = json.load(json_file)

# Ahora tienes el contenido del archivo JSON en la variable 'annotations_data'

# Ver la estructura general de las anotaciones
print("Estructura general de las anotaciones:")
print(annotations_data.keys())  # Esto mostrará las claves principales en el archivo JSON

# Analizar las anotaciones específicas
annotations = annotations_data["annotations"]

# Por ejemplo, puedes imprimir las primeras 5 anotaciones
print("\nPrimeras 5 anotaciones:")
for annotation in annotations[:5]:
    print(annotation)
```
